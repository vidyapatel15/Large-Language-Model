```python
import pyttsx3

# Initialize the TTS engine
engine = pyttsx3.init()

# Small Fine-Tuning: Adjust rate (speed) and voice properties
rate = engine.getProperty('rate')     # Default is usually 200
engine.setProperty('rate', 150)       # Decrease speaking speed

volume = engine.getProperty('volume') # Volume level (0.0 to 1.0)
engine.setProperty('volume', 0.9)     # Slightly reduce volume

voices = engine.getProperty('voices') # List of available voices
engine.setProperty('voice', voices[0].id)  # Change voice (try 1 for female if available)

# Text input
text = "Hello! I hope you're enjoying learning AI. This is a text-to-speech example in Python."

# Speak the text
engine.say(text)
engine.runAndWait()
```

```python
from gtts import gTTS
import os

text = "A new take on Aesop’s classic fable, and one of the most memorable stories for kids. When a shepherd’s youngest son is asked watch a flock of sheep, he can’t help but find it breathtakingly booooooooooring. So to drum up a little “fun”, he decides to shout for the villagers to save him from a wolf hiding in a tree"

# Language: English
tts = gTTS(text=text, lang='en')

# Save audio to file
tts.save("output.mp3")

# Play audio (Windows)
os.system("start output.mp3")
```

```python
# Import necessary modules
from gtts import gTTS
from IPython.display import Audio

# ==== Input Text ====
text = "Hello! This is a fine-tuned text-to-speech demo using Google TTS."

# ==== Fine-tuning options ====
language = 'en'    # Language code: 'en' = English, 'hi' = Hindi, 'ko' = Korean, etc.
slow = False       # Set to True to slow down speech

# ==== Generate and save speech ====
tts = gTTS(text=text, lang=language, slow=slow)
tts.save("output.mp3")

# ==== Play the speech ====
Audio("output.mp3")
```
